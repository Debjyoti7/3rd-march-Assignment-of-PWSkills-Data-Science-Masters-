# Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

# Here are five plots that we can plot using the Seaborn library, along with their uses:
# Histogram Plot: A histogram is a graphical representation of the distribution of numerical data. It shows the frequency distribution of a continuous variable. It is useful for visualizing the shape of the distribution, identifying outliers, and checking for symmetry or skewness.
# Scatter Plot: A scatter plot is a two-dimensional plot that uses dots to represent data points. It is useful for visualizing the relationship between two variables. It can help identify patterns or trends in the data and detect outliers.
# Line Plot: A line plot is a graphical representation of data points connected by straight line segments. It is useful for visualizing trends over time or other continuous variables. It can help identify patterns or trends in the data and detect outliers.
# Heatmap Plot: A heatmap is a graphical representation of data in which values are represented by colors. It is useful for visualizing the relationship between two or more variables. It can help identify patterns or trends in the data and detect outliers.
# Bar Plot: A bar plot is a graphical representation of data using rectangular bars. It is useful for comparing the values of different categories. It can help identify patterns or trends in the data and detect outliers.

# Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
fmri = sns.load_dataset("fmri")

sns.lineplot( x = 'timepoint' , y = 'signal'  , data = fmri , hue = 'region')
```

# Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.

```python
sns.scatterplot(x = 'pclass' , y = 'age' , data = titanic , hue = 'embark_town' )
```

# Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
diamonds = sns.load_dataset('diamonds')
sns.histplot(x = 'price' , data = diamonds , hue = 'color')
```

```python
sns.histplot(x = 'cut' , data = diamonds , hue = 'color')
```

# Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
iris = sns.load_dataset('iris')
sns.pairplot(data = iris , hue = 'species')
```

# Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
flights = sns.load_dataset('flights')
sns.heatmap(flights.corr() , annot = True , cmap= 'coolwarm') 
```
